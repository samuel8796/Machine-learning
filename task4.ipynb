```python
# Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
#col_names = ['feature_0', 'feature_1', 'feature_2', 'feature_3','feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8','feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13','feature_14', 'feature_15', 'feature_16', 'feature_17','feature_18', 'feature_19', 'feature_20', 'label']
# load dataset
df = pd.read_csv("introML2019F.task4.train.csv")#, header=None)#, names=col_names)
df_answer = pd.read_csv("introML2019F.task4.test.shuffled.noanswers.csv")
df.head()
```

```python
df.tail()
```

```python
#split dataset in features and target variable
feature_cols = ['feature_0', 'feature_1', 'feature_2', 'feature_3','feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8','feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13','feature_14', 'feature_15', 'feature_16', 'feature_17','feature_18', 'feature_19', 'feature_20', 'feature_21', 'feature_22','feature_23', 'feature_24', 'feature_25']
X = df[feature_cols] # Features
y = df.label # Target variable
x_answer = df_answer[feature_cols]
#print(x_answer)
print(y)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values= -1.00000 , strategy='mean')
imp_mean.fit(X)
X = imp_mean.transform(X)
X = pd.DataFrame(X,index=None,columns = ['feature_0', 'feature_1', 'feature_2', 'feature_3','feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8','feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13','feature_14', 'feature_15', 'feature_16', 'feature_17','feature_18', 'feature_19', 'feature_20', 'feature_21', 'feature_22','feature_23', 'feature_24', 'feature_25'])
print(X)
```

```python
#imp_mean = SimpleImputer(missing_values= -1.00000 , strategy='mean')
#imp_mean.fit(x_answer)
x_answer=imp_mean.transform(x_answer)
x_answer = pd.DataFrame(x_answer,index=None,columns = ['feature_0', 'feature_1', 'feature_2', 'feature_3','feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8','feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13','feature_14', 'feature_15', 'feature_16', 'feature_17','feature_18', 'feature_19', 'feature_20', 'feature_21', 'feature_22','feature_23', 'feature_24', 'feature_25'])
print(x_answer)
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = RandomForestClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
#print(clf.score(X_test,y_test))
#print(clf.feature_importances_)
```
